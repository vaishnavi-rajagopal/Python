```python
import numpy as np
import requests
import json
from geopy.distance import geodesic
from function_module import all_functions

# Get the place details for Kenya
op_query_place = """[out:json];
area["ISO3166-1"="KE"][admin_level=2];
(node["place"](area););out center;"""

data_place = all_functions.exec_resp(op_query_place)
place = all_functions.get_place(data_place,'place')

# Get the hospital, sanitation and water details for Kenya from a list of all amenities
op_query_amn = """[out:json];
area["ISO3166-1"="KE"][admin_level=2];
(
node["amenity"](area);
way["amenity"](area);
rel["amenity"](area);
);out center;"""

data_amn = all_functions.exec_resp(op_query_amn)
name = all_functions.get_name(data_amn)
name_hlth = []
name_sani = []
name_watr = []
hlth_amn = ['hospital','pharmacy','clinic','community chemist','ujamaa family centre','nursing_home',\
            'clinic;hospital','health_center','health']
sani_amn = ['toilets']#,'compost_site','waste_basket','waste_disposal','sanitary_dump_station']
watr_amn = ['water*','Water*','community_centre;drinking_water','drinking_water','pump house']
for i in range(len(name)):
    if name[i]['amenity'] in hlth_amn:
        name_hlth.append(name[i])
    elif name[i]['amenity'] in sani_amn:
        name_sani.append(name[i])
    elif name[i]['amenity'] in watr_amn:
        name_watr.append(name[i])
with open('data_amen_hlth.json', 'w') as outfile:
    json.dump(name_hlth, outfile)
with open('data_amen_sani.json', 'w') as outfile:
    json.dump(name_sani, outfile)
with open('data_amen_watr.json', 'w') as outfile:
    json.dump(name_watr, outfile)
    
# Get the hospital, sanitation and water details for Kenya from a list of all amenities
op_query_bldg = """[out:json];
area["ISO3166-1"="KE"][admin_level=2];
(
node["building"](area);
way["building"](area);
rel["building"](area);
);out center;"""

data_bldg = all_functions.exec_resp(op_query_bldg)
#name_bldg = all_functions.get_name(data_bldg)
    
with open('data_bldg.json', 'w') as outfile:
    json.dump(data_bldg, outfile)
```
